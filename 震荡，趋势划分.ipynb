```python
from KRData.HKData import HKFuture
from pyecharts import Kline, Line, Overlap
import numpy as np
import pandas as pd
hf = HKFuture('krdata', 'kairuitouzi')
data=hf.get_bars('HSI1807', start='20180705', end='20180705')
```

```python
def kline_tooltip(params):
    # s = params[0].seriesName + '</br>'
    d = params[0].name + '</br>'
    o = ': ' + params[0].data[1] + '</br>'
    h = ': ' + params[0].data[2] + '</br>'
    l = ': ' + params[0].data[3] +  '</br>'
    c = ': ' + params[0].data[4] + '</br>'
    text = d + o + h + l + c
    return text
kline = Kline('K')
kline.add('kline', data['datetime'].astype(str).values, 
          data[['open', 'close', 'low', 'high']].values, 
          tooltip_trigger='axis',
          tooltip_formatter=kline_tooltip,
          is_datazoom_show=True,
          datazoom_type='both')
```

```python
close = data['close']
close_y = close.shift(1)
close_t = close.shift(-1)
extreme_max_point = close[((close>close_y)&(close>close_t))]
extreme_min_point = close[((close<close_y)&(close<close_t))]
extreme_point = close[((close>close_y)&(close>close_t))|((close<close_y)&(close<close_t))]
```

```python
close_ = Line('close')
close_.add('close', data['datetime'].astype(str).values, data['close'].values, yaxis_min='dataMin', yaxis_max='dataMax',is_datazoom_show=True,
          datazoom_type='both')
```

```python
l = []
d = []
for i in extreme_point.items():
    if len(d)>=2:            
        _min = min(v[1] for v in d)
        _max = max(v[1] for v in d)
        avg = sum(v[1] for v in d)/len(d)
        vib = _max - _min
        if i[1] > _max:
            if (i[1] - 0.5*vib - avg) > 0.618*vib:
                l.append(d.copy())
                d.clear()
        elif i[1] < _min:
            if (avg- 0.5*vib - i[1]) > 0.618*vib:
                l.append(d.copy())
                d.clear()
    d.append(i)
l
```

```python
l = []
d = []
for i in extreme_point.items():
    if len(d)>=2:            
        avg = np.average([v[1] for v in d])
        std = np.std([v[1] for v in d])
        if i[1] > avg:
            if i[1] > avg + 2*std:
                l.append(d.copy())
                d.clear()
        elif i[1] < avg:
            if i[1] > avg + 2*std:
                l.append(d.copy())
                d.clear()
    d.append(i)
l
```

$$ E=\frac{\left| P_{n}-P_{1} \right|}{\sum_{i=2}^n{\left| P_{i}-P_{i-1} \right|}} $$

```python
n=30
def efficiency(c):
    Offset = abs(c[0] - c[-1])
    l = pd.Series(c)
    Distance = (l-l.shift(1)).abs().sum()
    e = Offset/Distance
    return e
    
E = close.rolling(n).apply(efficiency).fillna(0)
# E_MA = E.rolling(10).mean().fillna(0)
```

```python
data = hf.get_main_contract_bars('HSI', start='20180701')
grouped = data.groupby('trade_date')
Threshold = 0.3
n = 30  # 
vk = 30  # 
rk = 10  # 
```

```python
def spilt(df):
    def efficiency(c):
        Offset = abs(c[0] - c[-1])
        l = pd.Series(c)
        Distance = (l-l.shift(1)).abs().sum()
        e = Offset/Distance
        return e
    
    close = df['close']
    E = close.rolling(n).apply(efficiency).fillna(0)
    
    v = []
    r = []
    temp=[]
    period=[]
    rcount = 0
    vcount = 0
    for d, p in E.iteritems():
        temp.append((d, p))
        if p >= Threshold:
            if vcount >= vk: 
                period.append(('v', (temp[0][0], temp[-1][0])))
                temp.clear()
            elif vcount !=0:
                temp.clear()
            rcount += 1
            vcount = 0
        else:
            if rcount >= rk:
                period.append(('r', (temp[0][0], temp[-1][0])))
                temp.clear()
            elif rcount != 0:
                temp.clear()
            vcount += 1     
            rcount = 0
    return period
           
```

```python
ret = grouped.apply(spilt)
```

```python
date='20180716'
df = grouped.get_group(date)  # 
def efficiency(c):
    Offset = abs(c[0] - c[-1])
    l = pd.Series(c)
    Distance = (l-l.shift(1)).abs().sum()
    e = Offset/Distance
    return e

close = df['close']
E = close.rolling(n).apply(efficiency).fillna(0)

def kline_tooltip(params):
    # s = params[0].seriesName + '</br>'
    d = params[0].name + '</br>'
    o = ': ' + params[0].data[1] + '</br>'
    h = ': ' + params[0].data[2] + '</br>'
    l = ': ' + params[0].data[3] +  '</br>'
    c = ': ' + params[0].data[4] + '</br>'
    text = d + o + h + l + c
    return text
kline = Kline(f'K{date}')
kline.add('kline', df['datetime'].astype(str).values, 
          df[['open', 'close', 'low', 'high']].values, 
          tooltip_trigger='axis',
          tooltip_formatter=kline_tooltip,
          is_datazoom_show=True,
          datazoom_type='both')

E_ = Line('efficiency')
E_.add('efficiency', E.index.astype(str).values, E.values, yaxis_min='dataMin', yaxis_max='dataMax',is_datazoom_show=True,
          datazoom_type='both', line_color='blue')

overlap = Overlap('')
overlap.add(kline, yaxis_index=0)
overlap.add(E_, yaxis_index=1, is_add_yaxis=True)
overlap
```
